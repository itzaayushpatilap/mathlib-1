```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 7])
x = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
 
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot 4')
plt.tight_layout()

# Show the plots
plt.show()
    
```
